# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(autoload_with=engine)
Station = Base.classes.station
Measurement = Base.classes.measurement
```

```python
# View all of the classes that automap found
classes = Base.classes.keys()
class_list = list(classes)
output = str(class_list)
print(output)
```

```python
# Save references to each table
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
Measurement = Base.classes.measurement
```

# Exploratory Precipitation Analysis

```python
# Create a session to link Python to the database
session = Session(engine)

# Find the most recent date in the data set
most_recent_date = session.query(func.max(Measurement.date)).scalar()
formatted_date = f"('{most_recent_date}',)"
print(f"The most recent date in the data set is: {formatted_date}")
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Create a session to link Python to the database
session = Session(engine)

# Find the most recent date in the data set
most_recent_date = session.query(func.max(Measurement.date)).scalar()

# Calculate the date one year from the last date in the data set
one_year_ago = dt.datetime.strptime(most_recent_date, '%Y-%m-%d') - dt.timedelta(days=366)

# Perform a query to retrieve the data and precipitation scores for the last 12 months
results = session.query(Measurement.date, Measurement.prcp).filter(Measurement.date >= one_year_ago).all()

# Save the query results as a Pandas DataFrame and set column names
df = pd.DataFrame(results, columns=['Date', 'Precipitation'])

# Sort the DataFrame by date
df = df.sort_values('Date')
# Use Pandas Plotting with Matplotlib to plot the data
df.plot(x='Date', y='Precipitation', figsize=(6, 4), title='Precipitation in Honolulu, Hawaii (Last 12 Months)')
plt.xlabel('Date')
plt.ylabel('Inches')
plt.xticks(rotation=90)
plt.legend(loc='upper right')
plt.show()
```

```python

# Use Pandas to calculate the summary statistics for the precipitation data
# Create a DataFrame from the results
df = pd.DataFrame(results, columns=['date', 'precipitation'])

# Calculate the summary statistics
summary_stats = df.describe()

# Print the summary statistics
print(summary_stats)
```

# Exploratory Station Analysis

```python
# Create a session to link Python to the database
session = Session(engine)

# Design a query to calculate the total number of stations
total_stations = session.query(func.count(Station.station)).scalar()

# Print the total number of stations
print(f"The total number of stations in the dataset is: [({total_stations},)]")
```

```python
# Create a session to link Python to the database
session = Session(engine)

# Design a query to find the most active stations and their row counts
station_counts = session.query(Measurement.station, func.count(Measurement.station)).\
                group_by(Measurement.station).\
                order_by(func.count(Measurement.station).desc()).all()

# Print the most active stations and their row counts vertically
output = []
for station, count in station_counts:
    output.append((station, count))

print(output)
```

```python
# Create a session to link Python to the database
session = Session(engine)

# Find the most active station
most_active_station = session.query(Measurement.station).\
                group_by(Measurement.station).\
                order_by(func.count(Measurement.station).desc()).first()[0]

# Calculate the lowest, highest, and average temperature for the most active station
temperature_stats = session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs)).\
                    filter(Measurement.station == most_active_station).all()

# Print the temperature statistics
print(temperature_stats)
```

```python
import matplotlib.pyplot as plt

# Create a session to link Python to the database
session = Session(engine)

# Query the last 12 months of temperature observation (TOBS) data for the most active station
most_active_station = session.query(Measurement.station).\
                        group_by(Measurement.station).\
                        order_by(func.count(Measurement.station).desc()).\
                        first()

most_active_station_tobs = session.query(Measurement.tobs).\
                            filter(Measurement.station == most_active_station[0]).\
                            filter(Measurement.date >= one_year_ago).all()

# Extract the temperatures from the query results
temperatures = [temp[0] for temp in most_active_station_tobs]

# Plot the results as a histogram with bins=12
plt.hist(temperatures, bins=12)
plt.xlabel('Temperature (°F)')
plt.ylabel('Frequency')
plt.title('Temperature Observation (TOBS) for Most Active Station')
plt.tight_layout()
plt.show()
```
